```python
(ds_nonuc_CC['HO2concAfterChem'].isel(lev = 0).mean(dim = 'time')-ds_normal_CC['HO2concAfterChem'].isel(lev = 0).mean(dim = 'time')).plot()
```

```python
fig, ax = plt.subplots(figsize = (10,6),subplot_kw={'projection':ccrs.LambertConformal()})
(ds_nonuc_CC['HO2concAfterChem'].isel(lev = 0).mean(dim = 'time')-ds_normal_CC['HO2concAfterChem'].isel(lev = 0).mean(dim = 'time')).plot( 
                                                                                                                     cmap=plt.cm.RdBu, 
                                                                                                                     transform=ccrs.PlateCarree() #fit data into map
)
ax.add_feature(cfeat.STATES)
ax.coastlines() #add coastlines
ax.set_extent([-120.0, -60.0, 20.0, 50.0])
```

```python
fig, ax = plt.subplots(figsize = (10,6),subplot_kw={'projection':ccrs.LambertConformal()})
(ds_nonuc_CC['OHconcAfterChem'].isel(lev = 0)-ds_normal_CC['OHconcAfterChem'].isel(lev = 0)).mean(dim = 'time').plot(vmin = -100000, 
                                                                                                                     vmax = 100000, 
                                                                                                                     center = 0, 
                                                                                                                     cmap=plt.cm.RdBu, 
                                                                                                                     transform=ccrs.PlateCarree() #fit data into map
)
ax.add_feature(cfeat.STATES)
ax.coastlines() #add coastlines
ax.set_extent([-120.0, -60.0, 20.0, 50.0])
```
